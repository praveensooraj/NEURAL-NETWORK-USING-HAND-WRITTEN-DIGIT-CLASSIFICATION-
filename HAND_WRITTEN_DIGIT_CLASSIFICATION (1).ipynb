```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
mnist=keras.datasets.mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
x_train,x_test= x_train/255.0,x_test/255.0
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[10])
```

```python
y_train[10]
```

```python
model=keras.Sequential([layers.Flatten(input_shape=(28,28)),layers.Dense(128,activation='relu'),layers.Dense(10,activation='softmax')])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
y_pred=model.predict(x_test)
```

```python
import numpy as np
y_pred1=np.argmax(y_pred,axis=1)
y_pred1[12]
```

```python
y_test[12]
```

```python
plt.imshow(x_test[12])
```

```python
plt.imshow(x_test[18],cmap='gray')
plt.title(f'y_test:{y_test[18]},y_pred:{y_pred1[18]}')
```

```python
import random
num_samples=10
selected_indices=random.sample(range(len(x_test)),num_samples)
for i in selected_indices:
  plt.figure()
  plt.imshow(x_test[i],cmap='gray')
  plt.title(f'True:{y_test[i]},predicted:{y_pred1[i]}')
  plt.show()
```
